```python
from collections import Counter
import re

# Italian stopwords: articles, prepositions, conjunctions, common function words
ITALIAN_STOPWORDS = {
    "il", "lo", "la", "i", "gli", "le", "l",
    "un", "uno", "una", "un",
    "di", "del", "dello", "della", "dei", "degli", "delle",
    "a", "al", "allo", "alla", "ai", "agli", "alle", "dell",
    "da", "dal", "dallo", "dalla", "dai", "dagli", "dalle",
    "in", "nel", "nello", "nella", "nei", "negli", "nelle",
    "su", "sul", "sullo", "sulla", "sui", "sugli", "sulle",
    "con", "col", "coi",
    "per", "tra", "fra",
    "e", "ed", "o", "od", "ma", "se", "che", "come", "anche",
    "non", "più", "già", "si", "ne", "ci", "vi",
    "è", "era", "sono", "fu", "ha", "hanno",
    "alla", "alle", "agli", "all",
    "questo", "questa", "questi", "queste",
    "suo", "sua", "suoi", "sue",
    "loro", "ogni", "altri", "altre",
}

# Load data
df = pd.read_json('ner_gemini_output.json')

# Tokenize: lowercase, split on non-alpha characters, filter stopwords and short tokens
all_words = []
for title in df['title_it'].dropna():
    tokens = re.split(r'[^a-zA-ZÀ-ÿ]+', title.lower())
    for token in tokens:
        if token and len(token) > 2 and token not in ITALIAN_STOPWORDS:
            all_words.append(token)

# Count and display top 20
top20 = Counter(all_words).most_common(20)

results = pd.DataFrame(top20, columns=['word', 'count'])
results.index += 1  # 1-based ranking
print(results.to_string())
```
